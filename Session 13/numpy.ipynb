```python
# NumPy 
import numpy as np
```

```python
# Create a numpy array from a list
arr = np.array([1, 2, 3, 4, 5, 6])
```

```python
# Views the array
arr
```

```python
# Type of array
# It is not a list
# It is a NumPy array (numpy.ndarray)
type(arr)
```

```python
# Make an array which
# Starts at 1
# Ends at 100 (not included)
# Interval 3
np.arange(1,100,3)
# This same concept can be used to make similar arrays
```

```python
np.arange(7, 7000, 153)
```

```python
# Make an array of 10 zeros
np.zeros(10)
```

```python
# Make an array of 10 ones
np.ones(10)
```

```python
# To make an array of another number (e.g. 5, use this way)
np.zeros(10) + 5
```

```python
# You can have other operations too (-, *, /)
np.ones(10) - 0.12756947
```

```python
np.ones(10) * 3.14
```

```python
np.ones(10) / 89
```

```python
np.ones(10) ** 4
```

```python
# Make 2 dimensional arrays 
np.zeros((5,5))
```

```python
np.ones((5,4))
```

```python
# np.eye makes an array with ones in a diagonal, zeros to fill the rest
# Parameter taken is the number of rows 
np.eye(5)
```

```python
# np.linspace returns an array with numbers equally spaced in the range (range is first to params)
# third parameter is the number of numbers
np.linspace(1,1000,50)
```

```python
# np.random.rand returns an array of random numbers between 0 and 1
# To get the values more than 1, use the operations
x = np.random.rand(5,5)
```

```python
x
```

```python
np.random.rand(5, 5) + 7
```

```python
# Get minimum
x.min()
```

```python
# Get max
x.max()
```

```python
# Indexing
x[0, 3]
```

```python
# Slicing
x[1:4,2:4]
```

```python
# Will change those 6 values above to 0.14159
x[1:4,2:4]= 0.14159  # Broadcasting
```

```python
x
```

```python
x[x>0.5] # Conditional Indexing 
# Getting all numbers in the array which are above 0.5
# All Python COndional Rules apply in the example above
# 'and', 'or', 'not', ==, >=, <=, <, >
```
